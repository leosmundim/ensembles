# Árvore 1

Modelar os resíduos

```python
r = np.array(df.r).reshape(1,-1)
x= np.array(df.x).reshape(1,-1)

model = DecisionTreeRegressor(max_depth = 2, 
                             ccp_alpha=0)
reg_tree_1 = model.fit(x.T, r.T)
```

```python
df['p1'] = reg_tree_1.predict(x.T) #predito da árvore 1
df['P1'] = df.p + df.p1 # Predito do boosting 1 (acumulado) - preditiva final 
df['r1'] = df.r - df.p1 # resíduos do boosting 1
```

### Comparativo Valores Preditos vs Observados e Erro Boosing 1

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

sns.scatterplot(x='x', y='y', hue='r', palette='viridis', data=df, ax=ax1)
sns.lineplot(x='x', y='p', data=df, color='black', label='preditos', ax=ax1)
ax1.set_title('Valores Observados vs Preditos Inicial')

sns.scatterplot(x='x', y='r', hue='r1', palette='viridis', data=df, ax=ax2)
sns.lineplot(x='x', y='p1', data=df, color='black', label='preditos', ax=ax2)
ax2.set_title('Gráfico de Resíduos vs Preditos Boosting 1 ')

sns.scatterplot(x='x', y='y', hue='r1', palette='viridis', data=df, ax=ax3)
sns.lineplot(x='x', y='P1', data=df, color='black', label='preditos', ax=ax3)
ax3.set_title('Valores Observados vs Preditos Boosting 1')
```

## Comparativo dos Erros

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

sns.scatterplot(x='x', y='r', hue='r', palette='viridis', data=df, ax=ax1)
sns.lineplot(x='x', y='p1', data=df, color='black', label='preditos boosing 1', ax=ax1)
ax1.set_title('Gráfico de Resíduos Sem Boosting')

sns.scatterplot(x='x', y='y', hue='r1', palette='viridis', data=df, ax=ax2)
sns.lineplot(x='x', y='P1', data=df, color='black', label='p + p1', ax=ax2)
ax2.set_title('Valores Observados vs Preditos Boosting 1')

sns.scatterplot(x='x', y='r1', hue='r1', palette='viridis', data=df, ax=ax3)
ax3.set_title('Gráfico de Resíduos Boosting 1')

```

## Árvore 2 - Boosting 2

Modelar os Resíduos do Boosting 1

```python
r1 = np.array(df.r1).reshape(1,-1)
x= np.array(df.x).reshape(1,-1)

model = DecisionTreeRegressor(max_depth = 2, 
                             ccp_alpha=0)
reg_tree_2 = model.fit(x.T, r1.T)
```

```python
df['p2'] = reg_tree_2.predict(x.T) #predito da árvore 2
df['P2'] = df.P1 + df.p2 # Predito do boosting 2 (acumulado) - preditiva final 
df['r2'] = df.r1 - df.p2 # resíduos do boosting 2
```

### Comparativo Valores Preditos vs Observados e Erro Boosing 2

A tendência dos resíduos fique com poucos pontos onde o ajuste seja grande. No exemplo seria o lado esquerdo do gráfico de resíduos

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

sns.scatterplot(x='x', y='y', hue='r1', palette='viridis', data=df, ax=ax1)
sns.lineplot(x='x', y='P1', data=df, color='black', label='resíduos preditos', ax=ax1)
ax1.set_title('Valores Observados vs Preditos Boosting 1')

sns.scatterplot(x='x', y='r1', hue='r1', palette='viridis', data=df, ax=ax2)
sns.lineplot(x='x', y='p2', data=df, color='black', label='preditos', ax=ax2)
ax2.set_title('Gráfico de Resíduos Boosting 2')

sns.scatterplot(x='x', y='y', hue='r2', palette='viridis', data=df, ax=ax3)
sns.lineplot(x='x', y='P2', data=df, color='black', label='preditos', ax=ax3)
ax3.set_title('Valores Observados vs Preditos Boosting 2')
```

## Modelo Final

Poderíamos seguir melhorando o modelo com mais boostings
